```python
class node:
    def __init__(self,data):
        self.data=data
        self.prev=None
        self.next=None
class CDLL:
    def __init__(self):
        self.head=None
        self.tail=None
#Inserting portion:-
    def insert_at_begin(self,data):
        NB=node(data)
        NB.next=self.head
        self.head.prev=NB
        NB.prev=self.tail
        self.head=NB
        self.tail.next=NB
    def insert_at_end(self,data):
        NE=node(data)
        self.tail.next=NE
        NE.prev=self.tail
        NE.next=self.head
        self.tail=NE
        self.head.prev=self.tail   
    def insert_at_position(self,data,pos):          
        NP=node(data)                               
        temp=self.head
        for i in range(pos-1):
            temp=temp.next
        NP.prev=temp
        NP.next=temp.next
        temp.next.prev=NP
        temp.next=NP
#Deleting portion:-
    def del_at_begin(self):
        self.head.next.prev=self.tail
        self.head=self.head.next
        self.tail.next=self.head
    def del_at_end(self):
        self.head.prev=self.tail.prev
        self.tail.prev.next=self.head
        self.tail=self.tail.prev
    def del_at_position(self,pos):       
        temp=self.head                   
        for i in range(pos-1):
            temp=temp.next
        temp.next.next.prev=temp
        temp.next=temp.next.next
#Sorting portion(low to high):-
    def sort(self):
        L=[]
        temp=self.head
        L.append(temp.data)
        while temp.next!=self.head:
            temp=temp.next
            L.append(temp.data)
        L.sort()
        for i in range(len(L)):
            if i!=(len(L)-1):
                print(L[i],end='->')
        print(L[i])
    #Sorting portion(high to low):- 
    def decending_order_sort(self):
        L=[]
        temp=self.head
        L.append(temp.data)
        while temp.next!=self.head:
            temp=temp.next
            L.append(temp.data)
        for i in range(len(L)):
            for j in range(i+1,len(L)):
                if L[i]<=L[j]:
                    L[i],L[j]=L[j],L[i]
        for i in range(len(L)):
            if i!=(len(L)-1):
                print(L[i],end='->')
        print(L[i])
        
#Searching:-
    def searching(self,seach):
        L=[]
        temp=self.head
        L.append(temp.data)
        while temp.next!=self.head:
            temp=temp.next
            L.append(temp.data)
        for i in range(len(L)):
            if seach==L[i]:
                print(f"{seach} found at {i}")
                break
        else:
            print("**NOT FOUND**")
            
#Updating:-
    def updating(self,old,new):
        temp=self.head
        if temp.data==old:
            temp.data=new
        else:
            while temp.next!=self.head:
                temp=temp.next
                if temp.data==old:
                    temp.data=new
                    break
            else:
                print("**NOT FOUND**")
            
#Displaying portion:-        
    def display(self):
        if self.head ==None:
            print("***EMPTY LINKED LIST***")
        else:
            temp=self.head
            print(temp.data,end='->')
            while temp.next!=self.head:
                temp=temp.next
                print(temp.data,end='->')
            temp=temp.next
            print(temp.data)
    def display_from_right(self):
        if self.tail==None:
            print("***EMPTY LINKED LIST***")
        else:
            temp=self.tail
            print(temp.data,end='->')
            while temp.prev!=self.tail:
                temp=temp.prev
                print(temp.data,end='->')
                
            temp=temp.prev
            print(temp.data)
L=CDLL()
print('**************1st Node***************')
N1=node(10)
L.head=N1
L.tail=N1
N1.next=L.head
N1.prev=L.tail
print('Display from left side:',end='')
L.display()
print('Display from right side:',end='')
L.display_from_right()
print()
print('**************2nd Node***************')
N2=node(20)
N1.next=N2
N2.prev=N1
N2.next=L.head
L.tail=N2
L.head.prev=L.tail
print('Display from left side:',end='')
L.display()
print('Display from right side:',end='')
L.display_from_right()
print()
print('**************3rd Node***************')
N3=node(30)
N2.next=N3
N3.prev=N2
L.tail=N3
N3.next=L.head
L.head.prev=L.tail
print('Display from left side:',end='')
L.display()
print('Display from right side:',end='')
L.display_from_right()
print()
print('**************4th Node***************')
N4=node(40)
N3.next=N4
N4.prev=N3
L.tail=N4
N4.next=L.head
L.head.prev=N4
print('Display from left side:',end='')
L.display()
print('Display from right side:',end='')
L.display_from_right()
print()
print('**************5th Node***************')
N5=node(50)
N4.next=N5
N5.prev=N4
L.tail=N5
N5.next=L.head
L.head.prev=L.tail
print('Display from left side:',end='')
L.display()
print('Display from right side:',end='')
L.display_from_right()
print()

print('**************Adding Node at begin***************')
L.insert_at_begin(1001)
print('Display from left side:',end='')
L.display()
print('Display from right side:',end='')
L.display_from_right()
print()

print('**************Adding Node at end***************')
L.insert_at_end(11111)
print('Display from left side:',end='')
L.display()
print('Display from right side:',end='')
L.display_from_right()
print()

print('**************Adding Node at specific position***************')
L.insert_at_position(666,2)
print('Display from left side:',end='')
L.display()
print('Display from right side:',end='')
L.display_from_right()
print()

print('**************deleting Node at begin***************')
L.del_at_begin()
print('Display from left side:',end='')
L.display()
print('Display from right side:',end='')
L.display_from_right()
print()

print('**************deleting Node at end***************')
L.del_at_end()
print('Display from left side:',end='')
L.display()
print('Display from right side:',end='')
L.display_from_right()
print()

print('**************deleting Node at specific position***************')
L.del_at_position(2)
print('Display from left side:',end='')
L.display()
print('Display from right side:',end='')
L.display_from_right()
print()

print('**************SORTING(low to high)***************')
L.sort()
print()

print('**************SORTING(high to low)***************')
L.decending_order_sort()
print()

print('**************Searching***************')
try:
    L.searching(int(input("Enter a element to search:")))
except ValueError:
    print("***INVALID***")
    
print('**************Updating***************')
try:
    L.updating(int(input("Enter a old element:")),int(input("Enter a update element:")))
except ValueError:
    print("***INVALID***")
else:
    print('Display from left side:',end='')
    L.display()
    print('Display from right side:',end='')
    L.display_from_right()
    print()
finally:
    print("******************************************THE END*********************************************")
```
